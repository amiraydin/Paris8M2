Exercise 1: Simple Polynomial Plot



```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array x with 100 linearly spaced points between -10 and 10
x = np.linspace(-10, 10, 100)

# Create the array y for the polynomial function
y = 2 * x**3 - 5 * x**2 + 3 * x - 7

# Plot y versus x with a blue line
plt.plot(x, y, color='blue')

# Add title and labels
plt.title("Polynomial Function: $y = 2x^3 - 5x^2 + 3x - 7$")
plt.xlabel("x")
plt.ylabel("y")

# Show the grid
plt.grid()

# Display the plot
plt.show()
```

Custumization


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y = 2 * x**3 - 5 * x**2 + 3 * x - 7

# figure size to 10x6 inches
plt.figure(figsize=(10, 6))

# Plot y versus x with a blue line
plt.plot(x, y, color='blue')
# Add title and labels
plt.title("Polynomial Function: $y = 2x^3 - 5x^2 + 3x - 7$", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
# Show the grid
plt.grid()
# Display the plot
plt.show()
```
