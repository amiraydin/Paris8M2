Exercise 1: Simple Polynomial Plot



```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array x with 100 linearly spaced points between -10 and 10
x = np.linspace(-10, 10, 100)

# Create the array y for the polynomial function
y = 2 * x**3 - 5 * x**2 + 3 * x - 7

# Plot y versus x with a blue line
plt.plot(x, y, color='blue')

# Add title and labels
plt.title("Polynomial Function: $y = 2x^3 - 5x^2 + 3x - 7$")
plt.xlabel("x")
plt.ylabel("y")

# Show the grid
plt.grid()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y = 2 * x**3 - 5 * x**2 + 3 * x - 7

# figure size to 10x6 inches
plt.figure(figsize=(10, 6))

# Plot y versus x with a blue line
plt.plot(x, y, color='blue')
# Add title and labels
plt.title("Polynomial Function: $y = 2x^3 - 5x^2 + 3x - 7$", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
# Show the grid
plt.grid()
# Display the plot
plt.show()
```

```python
# Create an array x with 500 linearly spaced points between 0.1 and 10
x = np.linspace(0.1, 10, 500)

# Compute y1 = exp(x) and y2 = log(x)
y1 = np.exp(x)
y2 = np.log(x)

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot both functions
plt.plot(x, y1, color='blue', linestyle='-', label='y = exp(x)')
plt.plot(x, y2, color='orange', linestyle='--', label='y = log(x)')

plt.grid()
plt.title("Exponential and Logarithmic Functions", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)

# Add a legend
plt.legend()
# Save the plot as a PNG file with 100 DPI resolution
plt.savefig("exponential_logarithmic_functions.png", dpi=100)
plt.show()
```

```python
# figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Define the range for the first subplot
x1 = np.linspace(-2 * np.pi, 2 * np.pi, 500)

# Plot y = tan(x) and y = arctan(x) in the first subplot
ax1.plot(x1, np.tan(x1), label='y = tan(x)', color='blue')
ax1.plot(x1, np.arctan(x1), label='y = arctan(x)', color='orange')
ax1.set_title("Tangent and Arctangent Functions", fontsize=14)
ax1.set_xlabel("x", fontsize=12)
ax1.set_ylabel("y", fontsize=12)
ax1.legend()
ax1.grid()
ax1.set_ylim(-10, 10)  # Limit y-axis to avoid large tan values

# Define the range for the second subplot
x2 = np.linspace(-2, 2, 500)

# Plot y = sinh(x) and y = cosh(x) in the second subplot
ax2.plot(x2, np.sinh(x2), label='y = sinh(x)', color='green')
ax2.plot(x2, np.cosh(x2), label='y = cosh(x)', color='red')
ax2.set_title("Hyperbolic Sine and Cosine Functions", fontsize=14)
ax2.set_xlabel("x", fontsize=12)
ax2.set_ylabel("y", fontsize=12)
ax2.legend()
ax2.grid()

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```


    Plot a Contour Plot of a Mathematical Function:
        Create an array x of 200 points linearly spaced between -5 and 5, and a similar array y.
        Create a 2D grid with numpy.meshgrid and define the function f(x, y) = sin(sqrt(x^2 + y^2)).
        Plot a contour plot of this function.
    Customization:
        Add labels to the contours and use a different colormap.


```python
import numpy as np
import matplotlib.pyplot as plt

# Create arrays x and y with 200 points linearly spaced between -5 and 5
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)

# Create a 2D grid
X, Y = np.meshgrid(x, y)

# Define the function f(x, y) = sin(sqrt(x^2 + y^2))
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create the contour plot
plt.figure(figsize=(10, 6))
# Use a different colormap
contour = plt.contour(X, Y, Z, levels=20, cmap='plasma')

# Add labels to the contours
plt.clabel(contour, inline=True, fontsize=8)

# Add titles and labels
plt.title("Contour Plot of $f(x, y) = \sin(\sqrt{x^2 + y^2})$", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)

plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create arrays X and Y
x = np.arange(-5, 5.25, 0.25)
y = np.arange(-5, 5.25, 0.25)
X, Y = np.meshgrid(x, y)
# Compute Z
Z = np.cos(np.sqrt(X**2 + Y**2))
# Create the 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
# Customize the viewing angle
ax.view_init(elev=30, azim=45)
# Add color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of Z = cos(sqrt(X^2 + Y^2))')
plt.show()
```

```python
# arrays X, Y, and Z with values ranging from -5 to 5
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
z = np.linspace(-5, 5, 10)

# a meshgrid for X, Y, Z
X, Y, Z = np.meshgrid(x, y, z)
# Compute U, V, W
U = -Y
V = X
W = Z
# a figure
fig = plt.figure(figsize=(12, 8))

# a 3D vector field plot
ax = fig.add_subplot(111, projection='3d')

# Plot the vector field using quiver
# Here we scale the arrows for better visualization
scale = 5  # Adjust to change the length of the vectors
ax.quiver(X, Y, Z, U, V, W, length=0.5, color='blue', normalize=True)

ax.set_title('3D Vector Field: U = -Y, V = X, W = Z')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
from scipy.stats import gaussian_kde

# Generate 1000 random points in 3D space
x = np.random.normal(loc=0, scale=1, size=1000)
y = np.random.normal(loc=0, scale=1, size=1000)
z = np.random.normal(loc=0, scale=1, size=1000)

# 3D kernel density estimation (KDE)
xyz = np.vstack([x, y, z])
kde = gaussian_kde(xyz)

# grid of points to evaluate the KDE
xi, yi, zi = np.mgrid[-3:3:30j, -3:3:30j, -3:3:30j]
density = kde(np.vstack([xi.ravel(), yi.ravel(), zi.ravel()]))
density = density.reshape(xi.shape)

# figure
fig = plt.figure(figsize=(10, 7))

# 3D scatter plot of the original points
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o', alpha=0.1)  # Scatter plot with low transparency

# density scatter plot using the computed density
# Normalize the density for coloring
norm_density = (density - np.min(density)) / (np.max(density) - np.min(density))
ax.scatter(xi, yi, zi, c=norm_density.ravel(), cmap='viridis', alpha=0.5)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Density Plot of Random Points')
plt.show()
```
