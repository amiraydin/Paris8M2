```python
import keras
import tensorflow as tf
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print('X_train: ' + str(X_train.shape))
print('Y_train: ' + str(y_train.shape))
print('X_test:  '  + str(X_test.shape))
print('Y_test:  '  + str(y_test.shape))
```

```python
import matplotlib.pyplot as plt

for i in range(5):
  plt.subplot(330 + 1 + i)
  plt.imshow(X_train[i], cmap=plt.get_cmap('gray'))
  plt.show()
```

```python
X_train[0]
```

```python
axs, figs = plt.subplots(5, 5, figsize=(10, 12))
cp = 0
for i in range(5):
  for j in range(5):
    figs[i,j].imshow(X_train[cp], cmap="gray")
    #figs[i,j].set_title("label : {}".format(y_train[cp]))
    figs[i,j].set_title(f"label : {y_train[cp]}")
    cp += 1
```

```python
from keras import layers
import pandas as pd
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28,)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(10, activation='softmax'),
])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
model.evaluate(X_test, y_test)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import numpy as np
```

```python
print(
    "accuracy_score : ", accuracy_score(y_test, np.argmax(predictions, axis=-1)), "\n",
    #"f1_score : ", f1_score(y_test,  np.argmax(predictions, axis=-1)), "\n",
    #"recall_score : ", recall_score(y_test,  np.argmax(predictions, axis=-1)), "\n",
    #"precision_score : ", precision_score(y_test,  np.argmax(predictions, axis=-1)), "\n"
)
```

```python
model = keras.Sequential([
    keras.layers.Conv2D(64, 3, 3, input_shape=(28, 28,1)),
    keras.layers.MaxPooling2D(2, 2),
    keras.layers.Conv2D(128, 3, 3, activation="relu"),
    #keras.layers.MaxPooling2D(2, 2),

    keras.layers.Flatten(),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(10, activation='softmax'),
])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
predictions = model.predict(X_test)
print(classification_report(y_test, np.argmax(predictions, axis=-1)))
cm = confusion_matrix(y_test, np.argmax(predictions, axis=-1))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=range(10))
disp.plot()
plt.show()
```

```python
model.save("cnn_mnist_class.keras")
```
